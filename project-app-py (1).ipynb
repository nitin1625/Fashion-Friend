```python
import tensorflow
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import GlobalMaxPooling2D
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50,preprocess_input
```

```python
model=ResNet50(weights='imagenet',include_top=False,input_shape=(224,224,3))
```

```python
model.trainable=False
```

```python
model=tensorflow.keras.Sequential([model,GlobalMaxPooling2D()])
```

```python
print(model.summary())
```

```python
from  numpy.linalg import norm 
```

```python
def extract_features(img_path,model):
    img=image.load_img(img_path,target_size=(224,224))
    img_array=image.img_to_array(img)
    ex_arr=np.extend_dims(img_array,axis=0)
    pre_img=preprocessed_input( ex_arr)
    result=model.predict(pre_img).flatten()
    norm_result=result/norm(result)
    return norm_result
```

```python
def extract_features(img_path,model):
    img=image.load_img(img_path,target_size=(224,224))
    img_array=image.img_to_array(img)
    expanded_img_array=np.expand_dims(img_array,axis=0)
    preprocessed_img=preprocess_input(expanded_img_array)
    result=model.predict(preprocessed_img).flatten()
    normalized_result=result/norm(result)
    return normalized_result
```

```python
filenames=[]
for file in os.listdir('/kaggle/input/fashion-product-images-small/images'):
    filenames.append(os.path.join("/kaggle/input/fashion-product-images-small/images",file))
```

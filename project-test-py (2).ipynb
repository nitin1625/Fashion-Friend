```python
import pickle
import numpy as np 
import  tensorflow
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import ResNet50,preprocess_input
from tensorflow.keras.layers import GlobalMaxPooling2D
from numpy.linalg import norm
```

```python
feature_list=np.array(pickle.load(open('/kaggle/input/pickle/embeddings.pkl','rb')))
```

```python
filenames=pickle.load(open('/kaggle/input/pickle/filenames.pkl','rb'))
```

```python
model=ResNet50(weights='imagenet',include_top=False,input_shape=(224,224,3))
model.trainable=False
model=tensorflow.keras.Sequential([model,GlobalMaxPooling2D()])
```

```python
img_path='/kaggle/input/sample/1163.jpg'
```

```python
img=image.load_img(img_path,target_size=(224,224))
img_array=image.img_to_array(img)
ex_arr=np.expand_dims(img_array,axis=0)
pre_img=preprocess_input( ex_arr)
result=model.predict(pre_img).flatten()
norm_result=result/norm(result)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neighbors=NearestNeighbors(n_neighbors=6,algorithm='brute',metric='euclidean')
neighbors.fit(feature_list)
```

```python
distance,indices=neighbors.kneighbors([norm_result])
```

```python
print(indices)
```

```python
import cv2
```

```python
try:
    for file in indices[0]:
        temp_img=cv2.imread(filenames[file])
        cv2.imshow('output',temp_img)
    cv2.waitKey(0)
except Exception as e:
    print(e)
```
